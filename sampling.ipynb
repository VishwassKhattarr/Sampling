The dataset is highly imbalanced, which can bias machine learning models.

```python
df['Class'].value_counts()
```

```python
X = df.drop('Class', axis=1)
y = df['Class']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# StandardScaler is used to normalize the feature values.
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_bal, y_bal = smote.fit_resample(X_scaled, y)

pd.Series(y_bal).value_counts()

# SMOTE balances the dataset by generating synthetic samples for the minority class.
```

```python
from sklearn.model_selection import train_test_split

samples = []

for i in range(5):
    X_s, _, y_s, _ = train_test_split(
        X_bal, y_bal,
        train_size=0.7,
        random_state=i
    )
    samples.append((X_s, y_s))

len(samples)
```

```python
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import RandomOverSampler
from imblearn.combine import SMOTEENN
from imblearn.under_sampling import TomekLinks
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

models = {
    "M1": LogisticRegression(max_iter=1000),
    "M2": DecisionTreeClassifier(),
    "M3": RandomForestClassifier(),
    "M4": KNeighborsClassifier(),
    "M5": SVC()
}
```

```python
from sklearn.metrics import accuracy_score

sampling_methods = {
    "S1": RandomUnderSampler(),
    "S2": RandomOverSampler(),
    "S3": SMOTE(),
    "S4": TomekLinks(),
    "S5": SMOTEENN()
}

results = {}

for m_name, model in models.items():
    results[m_name] = {}
    for s_name, sampler in sampling_methods.items():
        X_res, y_res = sampler.fit_resample(samples[0][0], samples[0][1])
        model.fit(X_res, y_res)
        preds = model.predict(X_res)
        acc = accuracy_score(y_res, preds)
        results[m_name][s_name] = round(acc * 100, 2)
```

```python
results_df = pd.DataFrame(results).T
results_df
```
